```python
import pandas as pd

df = pd.read_excel("Smart_Grocery_Sales_Dataset_70Rows.xlsx")
df.head()
```

```python
# Total Sales
total_sales = df["Total_Amount"].sum()

# Average Order Value
average_order = df["Total_Amount"].mean()

print("Total Sales:", total_sales)
print("Average Order Value:", average_order)
```

```python
# Top Selling Products
top_products = df.groupby("Product")["Total_Amount"].sum().sort_values(ascending=False)

print("Top Selling Products")
print(top_products)
```

```python
# Category-wise Sales
category_sales = df.groupby("Category")["Total_Amount"].sum()

print("Category-wise Sales")
print(category_sales)
```

```python
# City-wise Sales
city_sales = df.groupby("City")["Total_Amount"].sum()

print("City-wise Sales")
print(city_sales)
```

```python
# Payment Method Analysis
payment_method = df["Payment_Method"].value_counts()

print("Payment Method Used")
print(payment_method)
```
